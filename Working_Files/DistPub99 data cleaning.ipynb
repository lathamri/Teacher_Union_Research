```python
def str_to_int(data):
    for col in data:
        j=0
        for i in data[col]:
            if str(i).isdigit():
                data.at[j,col] = int(i)
                j+=1
            else:
                pass
                j+=1
    return data

str_int_data = str_to_int(data_clean_bytes)

str_int_data
```

```python
def string_check(data):
    '''Checks for string values and returns columns which contain them.
    These must be excluded before running regressions.'''
    col_list = []
    for col in data:
        for i in data[col]:
            if isinstance(i, str):
                col_list.append(col)
    return set(col_list)

exc_list = string_check(str_int_data)
print(exc_list)
```

```python
# Split our data to let us predict instructer salaries

fin_data = str_int_data.drop(['StateAbb', 'StateAgencyID', 'CMSACode', 'SchoolYear-Fall', 'FiscalYear',
              'MetroCode'], axis=1)
```
