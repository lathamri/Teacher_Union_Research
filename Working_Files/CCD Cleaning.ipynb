```python
CCDdata = pd.read_sas('/Users/rileylatham/Downloads/Independent_Study/Data_Pool/Common Core data/sdf97d1a.sas7bdat',
                     encoding='unicode_escape')
```

```python
col_chg = {'Z32':'TotalSalaries', 'Z34':'TotalEmployeeBenefits'}

CCDdata = CCDdata.rename(columns=col_chg)

CCDdata = CCDdata[['LEAID', 'CENSUSID', 'FIPST', 'FIPSCO', 'CMSA', 'STNAME', 'STABBR', 'SCHLEV', 'AGCHRT',
                  'TOTALREV', 'TFEDREV', 'TSTREV', 'TLOCREV', 'TOTALEXP', 'TCURELSC', 'TCURINST', 
                  'TCURSSVC', 'TCUROTH', 'TNONELSE', 'TCAPOUT', 'TotalSalaries', 'TotalEmployeeBenefits']]
```

```python
SASSdata = pd.read_csv('/Users/rileylatham/Downloads/Independent_Study/Data_Pool/SASS data/Clean_SASS.csv')
```

```python
def str_to_int(data):
    for col in data:
        j=0
        for i in data[col]:
            if str(i).isdigit():
                data.at[j,col] = int(i)
                j+=1
            else:
                pass
                j+=1
    return data
```

```python
str_to_int = str_to_int(CCDdata)
str_to_int = str_to_int.dropna(axis=0)
```

```python
def LEAID_merge(SASS, CCD):
    j=186
    for val in CCD['LEAID']:
        CCD.at[j, 'LEAID'] = str(val)
        j+=1
    data = CCD[CCD['LEAID'].isin(SASS['CCDIDLEA'])]
    return data
```

```python
matched_data = LEAID_merge(SASSdata, str_to_int).dropna(axis=0)
```

```python
matched_data.to_csv(r'/Users/rileylatham/Downloads/Independent_Study/Data_Pool/Common Core data/Clean_CCD97.csv',
                    header=True, index=False)
```
